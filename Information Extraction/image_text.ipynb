# Imports and Set Up

```python
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
import re
from PIL import Image

plt.rcParams["figure.figsize"] = (10,10)
pytesseract.pytesseract.tesseract_cmd = 'E:/Program_Files/Tesseract/tesseract'
```

# Extraction

```python
def image_to_string(dir):
    text = pytesseract.image_to_string(dir)
    plt.imshow(Image.open(dir))
    print(f'TEST EXTRACTED:\n\n {text}')
    return text
```

```python
def image_to_txt(dir1, file_name):
    text = pytesseract.image_to_string(dir1)
    text_file = open(f'../Output Files/{file_name}.txt', 'w+')
    text_file.write(text)
    text_file.close()
```
